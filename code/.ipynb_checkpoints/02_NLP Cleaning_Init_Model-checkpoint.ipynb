#### Make everything lower case

```python
reviews_joined['comments_clean'] = reviews_joined['comments'].str.lower()
reviews_joined.head()
```

#### Created tokenizer and function to remove punctuation

```python
tokenizer = RegexpTokenizer(r'\b\w+\b')
```

```python
def remove_punct(text):
    tokens = tokenizer.tokenize(text)

    return ' '.join(tokens)
```

```python
reviews_joined['comments_clean'] = reviews_joined['comments_clean'].apply(remove_punct)
```

```python
pd.set_option('display.max_colwidth', 500)
reviews_joined[['comments', 'comments_clean']].sample(5)
```

#### There's 78,280 unique words in the cleaned comments

```python
results = set()
reviews_joined['comments_clean'].str.split().apply(results.update)
len(results)
```

#### Build initial model among those with a review score below 4

```python
X = reviews_joined['comments_clean']
X.shape
```

```python
y = reviews_joined['avg_score_below_4']
```

#### Need to stratify data because of imbalanced classes

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, stratify = y)
```

```python
X_train
```

#### Re-running with just Best_params to speed up run-time

```python
pipe_short = Pipeline([
    ('tf', TfidfVectorizer() ),
    ('logr', LogisticRegression() )
])
```

```python
pipe_params_short = {
    'tf__stop_words' : ['english'],
    'tf__max_features' : [5_000], #Using larger max features since this has over 300k comments
    'tf__min_df' : [2], # Minimum number of times a word needs to appear
    'tf__ngram_range' : [ (1,1) ]
}
```

```python
gs_short = GridSearchCV(pipe_short, pipe_params_short, n_jobs = -1, 
                  verbose = 2)
```

```python
gs_short.fit(X_train, y_train)
```

#### Good accuracy score doesn't mean much since most is in majority class

```python
gs_short.score(X_train, y_train), gs_short.score(X_test, y_test)
```

```python
recall_score(y_test, gs_short.predict(X_test) )
```

```python
f1_score(y_test, gs_short.predict(X_test) )
```

```python
ConfusionMatrixDisplay.from_estimator(gs_short, X_test, y_test);
```

#### Running a Grid Search over a TFIDF Vectorizer and Logistic Regression. 

#### Using TFIDF Vectorizer since I want to see what words show up more frequently in one class but less frequently in the other.

```python
pipe = Pipeline([
    ('tf', TfidfVectorizer() ),
    ('logr', LogisticRegression() )
])
```

```python
pipe_params = {
    'tf__stop_words' : ['english'],
    'tf__max_features' : [5_000, 10_000, 30_000, 60_000], #Using larger max features since this has over 300k comments
    'tf__min_df' : [2, 5, 10, 25], # Minimum number of times a word needs to appear
    'tf__ngram_range' : [ (1,1), (1,2), (1,3) ]
    # logr__class_weight
}
```

```python
gs = GridSearchCV(pipe, pipe_params, n_jobs = -1, 
                  verbose = 2)
```

```python
gs.fit(X_train, y_train)
```

```python
gs.best_score_
```

#### This model is actually WORSE than just predicting the majority class (0.999449)

```python
gs.best_score_ - reviews_joined['avg_score_below_4'].value_counts(normalize = True)[0]
```

#### The model takes awhile to run. In the future, best Params on this model are 5000 max features, 2 min_df, n_gram range of 1,1

```python
gs.best_params_
```

```python
gs.predict_proba(gs.best_estimator_.named_steps.tf.get_feature_names_out() )
```

```python
prob_df = pd.DataFrame(gs.predict_proba(gs.best_estimator_.named_steps.tf.get_feature_names_out() ), columns = ['above_4_rating', 'below_4_rating'])
prob_df
```

```python
feature_words = pd.DataFrame( gs.best_estimator_.named_steps.tf.get_feature_names_out(), columns = ['features'])
feature_words
```

#### Even the words that are most likely associated with a below 4 rating, are still very likely to be in the above_4_rating class

#### But some indication of what's driving poor ratings - reservation issues, bad/dirty and names of specific hosts (Todd, Louise, Stephen)

```python
word_probs = feature_words.join(prob_df)
word_probs.sort_values('below_4_rating', ascending = False).head(20)
```

#### Never predicts minority class
